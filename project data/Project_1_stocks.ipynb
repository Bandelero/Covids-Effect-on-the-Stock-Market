```python
def tick_counter(df):
    i=0
    df["tick"]=0
    size = int(df.shape[0])
    while i < size:
        df.at[i,"tick"]=i
        i+=1
```

```python
tick_counter(comm_s)
tick_counter(cons_d)
tick_counter(cons_s)
tick_counter(energy)
tick_counter(financ)
tick_counter(health)
tick_counter(indust)
tick_counter(it)
tick_counter(materi)
tick_counter(reales)
tick_counter(utilit)
```
