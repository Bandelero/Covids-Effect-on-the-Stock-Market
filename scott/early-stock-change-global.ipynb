```python
# data files
covid_file6 = "early2.csv"

# Read the data 
covid_stock_price6_df = pd.read_csv(covid_file6)

# Display the data table for preview
covid_stock_price6_df.head()
```

```python
covid_stock_price6_df
```

```python
from matplotlib.pyplot import figure

x = covid_stock_price6_df['Date']
a = covid_stock_price6_df['DOW_Change']
b = covid_stock_price6_df['NASDAQ_Change']
c = covid_stock_price6_df['SP500_Change']
d = covid_stock_price6_df['Total_Deaths']

figure(figsize=(16, 10))
plt.plot(x, b, color="blue", label="NASDAQ Change")
plt.plot(x, d, color="red", label="Covid Deaths")

plt.xticks(np.arange(0, len(x)+1, 10),rotation ='vertical')
plt.title("NASDAQ vs. Covid Deaths - Global")
plt.xlabel("Date")
plt.ylabel("Stock Price Change / Deaths")
plt.legend(loc="best")
#plt.savefig("covid-nasdaq5.png")
plt.show()
```

```python
# Plot the charts and apply some styling
figure(figsize=(16, 10))
plt.plot(x, c, color="black", label="S&P 500 Change")
plt.plot(x, d, color="red", label="Covid Deaths")

plt.xticks(np.arange(0, len(x)+1, 10),rotation ='vertical')
plt.title("S&P 500 vs. Covid Deaths - Global")
plt.xlabel("Date")
plt.ylabel("Stock Price Change / Deaths")
plt.xticks(rotation ='vertical')
plt.legend(loc="best")
#plt.savefig("covid-sp5004.png")
plt.show()
```

```python
# Plot the charts and apply some styling
figure(figsize=(16, 10))
plt.plot(x, a, color='green', label="DOW Change")
plt.plot(x, d, color="red", label="Covid Deaths")

plt.xticks(np.arange(0, len(x)+1, 10),rotation ='vertical')
plt.title("DOW vs. Covid Deaths - Global")
plt.xlabel("Date")
plt.ylabel("Stock Price Change / Deaths")
plt.xticks(rotation ='vertical')
plt.legend(loc="best")
#plt.savefig("covid-dow4.png")
plt.show()
```

```python
figure(figsize=(16, 10))
plt.plot(x, a, color='green', label="DOW")
plt.plot(x, b, color="blue", label="NASDAQ")
plt.plot(x, c, color="black", label="S&P 500")
plt.plot(x, d, color="red", label="Global Covid Deaths")

plt.xticks(np.arange(0, len(x)+1, 10),rotation ='vertical')
#plt.yticks(np.arange(0, max(y), 2))
plt.title("Stock Market Over Time")
plt.xlabel("Date")
plt.ylabel("Stock Price / Deaths")
plt.legend(loc="best")
#plt.savefig("covid-stocks4.png")
plt.show()
```

```python
correlation13 = st.pearsonr(b,d)
print(f"The correlation between Daily NASDAQ Change and Global Covid Deaths is {round(correlation13[0],2)}")
print(f"""The r-squared value is {round((correlation13[0]**2),2)}""")
```

```python
correlation14 = st.pearsonr(c,d)
print(f"The correlation between Daily S&P 500 Change and Global Covid Deaths is {round(correlation14[0],2)}")
print(f"""The r-squared value is {round((correlation14[0]**2),2)}""")
```

```python
(slope, intercept, rvalue, pvalue, stderr) = linregress(b, d)
regress_values = b * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
# Plot the linear regression model on top of the previous scatter plot.
plt.scatter(b,d, color="lightcoral", edgecolors="black")
plt.plot(b,regress_values,"r-")
#plt.annotate(line_eq,(7000,4000),fontsize=15,color="red")
plt.xlabel("NASDAQ Change")
plt.ylabel("Covid Deaths")
plt.title("Global Covid Deaths vs. NASDAQ")
#plt.savefig("")
plt.show()
print(f"The correlation between Daily NASDAQ Change and Global Covid Deaths is {round(correlation13[0],2)}")
print(f"""The r-squared value is {round((correlation13[0]**2),2)}""")
```
