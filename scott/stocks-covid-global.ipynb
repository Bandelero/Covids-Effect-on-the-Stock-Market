```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress
import os
import numpy as np

#covid_data3 = covid_data2.transpose()
#covid_data3.head()
#covid_data3.to_csv("covid_global.csv")

# data files
covid_file2 = "covid_global2.csv"

# Read the data 
covid_data2 = pd.read_csv(covid_file2)

# Display the data table for preview
covid_data2.head()
```

```python
nasdaq_file2 = "IXIC.csv"
nasdaq_data2 = pd.read_csv(nasdaq_file2)
nasdaq_data2.head()
```

```python
del nasdaq_data2['Open']
del nasdaq_data2['High']
del nasdaq_data2['Low']
del nasdaq_data2['Adj Close']
```

```python
dow_file2 = "DJIA.csv"
dow_data2 = pd.read_csv(dow_file2)
dow_data2.head()
```

```python
del dow_data2['Open']
del dow_data2['High']
del dow_data2['Low']
del dow_data2['Adj Close']
```

```python
dow_data2.head()
```

```python
sp500_file2 = "SP500.csv"
sp500_data2 = pd.read_csv(sp500_file2)
sp500_data2.head()
```

```python
del sp500_data2['Open']
del sp500_data2['High']
del sp500_data2['Low']
del sp500_data2['Adj Close']
```

```python
sp500_data2.head()
```

```python
sp500_data2.rename(columns={'Close': 'SP500_Close', 'Volume': 'SP500_Vol'}, inplace=True)
dow_data2.rename(columns={'Close': 'DOW_Close', 'Volume': 'DOW_Vol'}, inplace=True)
nasdaq_data2.rename(columns={'Close': 'NASDAQ_Close', 'Volume': 'NASDAQ_Vol'}, inplace=True)
```

```python
merged_data2_df = dow_data2.merge(sp500_data2, on = "Date")
```

```python
merged_data2_df
```

```python
all_stocks2_df = merged_data2_df.merge(nasdaq_data2, on = "Date")
```

```python
all_stocks2_df.head()
```

```python
covid_stocks2_df = all_stocks2_df.merge(covid_data2, on = "Date")
```

```python
covid_stocks2_df.head()
```

```python
covid_stock_price2_df = covid_stocks2_df.drop(['DOW_Vol', 'SP500_Vol', 'NASDAQ_Vol'], axis=1)
```

```python
covid_stock_price2_df.head()
```

```python
covid_stock_price2_df.to_csv("covid_stock_price_global.csv")
```

```python
from matplotlib.pyplot import figure

x = covid_stock_price2_df['Date']
a = covid_stock_price2_df['DOW_Close']
b = covid_stock_price2_df['NASDAQ_Close']
c = covid_stock_price2_df['SP500_Close']
d = covid_stock_price2_df['Total_Deaths']

figure(figsize=(16, 10))
plt.plot(x, b, color="blue", label="NASDAQ")
plt.plot(x, d, color="red", label="Covid Deaths")

plt.xticks(np.arange(0, len(x)+1, 10),rotation ='vertical')
plt.title("NASDAQ vs. Covid Deaths - Global")
plt.xlabel("Date")
plt.ylabel("Stock Price / Deaths")
plt.legend(loc="best")
plt.savefig("covid-nasdaq2.png")
plt.show()
```

```python
# Plot the charts and apply some styling
figure(figsize=(16, 10))
plt.plot(x, c, color="black", label="S&P 500")
plt.plot(x, d, color="red", label="Covid Deaths")

plt.xticks(np.arange(0, len(x)+1, 10),rotation ='vertical')
plt.title("S&P 500 vs. Covid Deaths - Global")
plt.xlabel("Date")
plt.ylabel("Stock Price / Deaths")
plt.xticks(rotation ='vertical')
plt.legend(loc="best")
plt.savefig("covid-sp5002.png")
plt.show()
```

```python
# Plot the charts and apply some styling
figure(figsize=(16, 10))
plt.plot(x, a, color='green', label="DOW")
plt.plot(x, d, color="red", label="Covid Deaths")

plt.xticks(np.arange(0, len(x)+1, 10),rotation ='vertical')
plt.title("DOW vs. Covid Deaths - Global")
plt.xlabel("Date")
plt.ylabel("Stock Price / Deaths")
plt.xticks(rotation ='vertical')
plt.legend(loc="best")
plt.savefig("covid-dow2.png")
plt.show()
```

```python
figure(figsize=(16, 10))
plt.plot(x, a, color='green', label="DOW")
plt.plot(x, b, color="blue", label="NASDAQ")
plt.plot(x, c, color="black", label="S&P 500")
plt.plot(x, d, color="red", label="Global Covid Deaths")

plt.xticks(np.arange(0, len(x)+1, 10),rotation ='vertical')
#plt.yticks(np.arange(0, max(y), 2))
plt.title("Stock Market Over Time")
plt.xlabel("Date")
plt.ylabel("Stock Price / Deaths")
plt.legend(loc="best")
plt.savefig("covid-stocks2.png")
plt.show()
```

```python
correlation4 = st.pearsonr(b,d)
print(f"The correlation between NASDAQ and Global Covid Deaths is {round(correlation4[0],2)}")
print(f"""The r-squared value is {round((correlation4[0]**2),2)}""")
```

```python
correlation5 = st.pearsonr(c,d)
print(f"The correlation between S&P 500 and Global Covid Deaths is {round(correlation5[0],2)}")
print(f"""The r-squared value is {round((correlation5[0]**2),2)}""")
```

```python
correlation6 = st.pearsonr(a,d)
print(f"The correlation between DOW and Global Covid Deaths is {round(correlation6[0],2)}")
print(f"""The r-squared value is {round((correlation6[0]**2),2)}""")
```

```python
(slope, intercept, rvalue, pvalue, stderr) = linregress(b, d)
regress_values = b * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
# Plot the linear regression model on top of the previous scatter plot.
plt.scatter(b,d, color="lightcoral", edgecolors="black")
plt.plot(b,regress_values,"r-")
plt.annotate(line_eq,(7000,4000),fontsize=15,color="red")
plt.xlabel("NASDAQ")
plt.ylabel("Covid Deaths")
plt.title("Global Covid Deaths vs. NASDAQ")
#plt.savefig("")
plt.show()
print(f"""The r-squared value is {round((correlation4[0]**2),2)}""")
```

```python
(slope, intercept, rvalue, pvalue, stderr) = linregress(c, d)
regress_values = c * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
# Plot the linear regression model on top of the previous scatter plot.
plt.scatter(c,d, color="lightcoral", edgecolors="black")
plt.plot(c,regress_values,"r-")
plt.annotate(line_eq,(2300,4000),fontsize=15,color="red")
plt.xlabel("S&P 500")
plt.ylabel("Covid Deaths")
plt.title("Global Covid Deaths vs. S&P 500")
#plt.savefig("")
plt.show()
print(f"""The r-squared value is {round((correlation5[0]**2),2)}""")
```
