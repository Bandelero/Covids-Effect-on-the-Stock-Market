```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress
import os
import numpy as np

#covid_data3 = covid_data2.transpose()
#covid_data3.head()
#covid_data3.to_csv("covid_global.csv")

# data files
covid_file4 = "covid_global2.csv"

# Read the data 
covid_data4 = pd.read_csv(covid_file4)

# Display the data table for preview
covid_data4.head()
```

```python
dow_file4 = "DJIA-change.csv"
dow_data4 = pd.read_csv(dow_file4)
dow_data4.head()
```

```python
sp500_file4 = "SP500-change.csv"
sp500_data4 = pd.read_csv(sp500_file4)
sp500_data4.head()
```

```python
del sp500_data4['Open']
del sp500_data4['High']
del sp500_data4['Low']
del sp500_data4['Adj Close']
del sp500_data4['Close']
```

```python
sp500_data4.rename(columns={'Change': 'SP500_Change', 'Volume': 'SP500_Vol'}, inplace=True)
dow_data4.rename(columns={'Change': 'DOW_Change', 'Volume': 'DOW_Vol'}, inplace=True)
nasdaq_data4.rename(columns={'Change': 'NASDAQ_Change', 'Volume': 'NASDAQ_Vol'}, inplace=True)
```

```python
merged_data4_df = dow_data4.merge(sp500_data4, on = "Date")
```

```python
merged_data4_df
```

```python
all_stocks4_df = merged_data4_df.merge(nasdaq_data4, on = "Date")
```

```python
all_stocks4_df.head()
```

```python
covid_stocks4_df = all_stocks4_df.merge(covid_data4, on = "Date")
```

```python
covid_stocks4_df.head()
```

```python
covid_stock_price4_df = covid_stocks4_df.drop(['DOW_Vol', 'SP500_Vol', 'NASDAQ_Vol'], axis=1)
```

```python
covid_stock_price4_df.head()
```

```python
#covid_stock_price4_df.to_csv("covid_stock_price_change_global.csv")
```

```python
from matplotlib.pyplot import figure

x = covid_stock_price4_df['Date']
a = covid_stock_price4_df['DOW_Change']
b = covid_stock_price4_df['NASDAQ_Change']
c = covid_stock_price4_df['SP500_Change']
d = covid_stock_price4_df['Total_Deaths']

figure(figsize=(16, 10))
plt.plot(x, b, color="blue", label="NASDAQ Change")
plt.plot(x, d, color="red", label="Covid Deaths")

plt.xticks(np.arange(0, len(x)+1, 10),rotation ='vertical')
plt.title("NASDAQ vs. Covid Deaths - Global")
plt.xlabel("Date")
plt.ylabel("Stock Price Change / Deaths")
plt.legend(loc="best")
#plt.savefig("covid-nasdaq4.png")
plt.show()
```

```python
# Plot the charts and apply some styling
figure(figsize=(16, 10))
plt.plot(x, c, color="black", label="S&P 500 Change")
plt.plot(x, d, color="red", label="Covid Deaths")

plt.xticks(np.arange(0, len(x)+1, 10),rotation ='vertical')
plt.title("S&P 500 vs. Covid Deaths - Global")
plt.xlabel("Date")
plt.ylabel("Stock Price Change / Deaths")
plt.xticks(rotation ='vertical')
plt.legend(loc="best")
#plt.savefig("covid-sp5004.png")
plt.show()
```

```python
# Plot the charts and apply some styling
figure(figsize=(16, 10))
plt.plot(x, a, color='green', label="DOW Change")
plt.plot(x, d, color="red", label="Covid Deaths")

plt.xticks(np.arange(0, len(x)+1, 10),rotation ='vertical')
plt.title("DOW vs. Covid Deaths - Global")
plt.xlabel("Date")
plt.ylabel("Stock Price Change / Deaths")
plt.xticks(rotation ='vertical')
plt.legend(loc="best")
#plt.savefig("covid-dow4.png")
plt.show()
```

```python
figure(figsize=(16, 10))
plt.plot(x, a, color='green', label="DOW")
plt.plot(x, b, color="blue", label="NASDAQ")
plt.plot(x, c, color="black", label="S&P 500")
plt.plot(x, d, color="red", label="Global Covid Deaths")

plt.xticks(np.arange(0, len(x)+1, 10),rotation ='vertical')
#plt.yticks(np.arange(0, max(y), 2))
plt.title("Stock Market Over Time")
plt.xlabel("Date")
plt.ylabel("Stock Price / Deaths")
plt.legend(loc="best")
#plt.savefig("covid-stocks4.png")
plt.show()
```

```python
correlation7 = st.pearsonr(b,d)
print(f"The correlation between Daily NASDAQ Change and Global Covid Deaths is {round(correlation7[0],2)}")
print(f"""The r-squared value is {round((correlation7[0]**2),2)}""")
```

```python
correlation8 = st.pearsonr(c,d)
print(f"The correlation between Daily S&P 500 Change and Global Covid Deaths is {round(correlation8[0],2)}")
print(f"""The r-squared value is {round((correlation8[0]**2),2)}""")
```

```python
(slope, intercept, rvalue, pvalue, stderr) = linregress(b, d)
regress_values = b * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
# Plot the linear regression model on top of the previous scatter plot.
plt.scatter(b,d, color="lightcoral", edgecolors="black")
plt.plot(b,regress_values,"r-")
#plt.annotate(line_eq,(7000,4000),fontsize=15,color="red")
plt.xlabel("NASDAQ Change")
plt.ylabel("Covid Deaths")
plt.title("Global Covid Deaths vs. NASDAQ")
#plt.savefig("")
plt.show()
print(f"The correlation between Daily NASDAQ Change and Global Covid Deaths is {round(correlation7[0],2)}")
print(f"""The r-squared value is {round((correlation7[0]**2),2)}""")
```
